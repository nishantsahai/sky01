```python
import datetime
import math


class RentedCars:
    '''This class generates all of all car available and car renting related information'''
    
    #This is car inventory. This can be extended based availability of the car. 
    #This car inventory can also be converted into function. For simiplicity of the program taking auto-populated car inventory list. 
    global car_list 
    car_list  = {1: ["Wagon R", 10, 150], 2: ["Maruti Swift", 8, 200], 3: ["Tata Nixon", 6, 200], 4 : ["Toyota Innova", 7, 250], 5: ["Toyota Fortuner", 3, 350]}
    
    #Contructor
    def __init__(self):
        pass
    
    #Displays all available cars to customer to book from. 
    def available_cars(self):         
        sum = 0    
        for values in car_list.values():
            car_name, available_quantity, hour_price = values
            sum = sum + available_quantity
        print('\033[1m' + "{:<5} {:<15} {:<15} {:<25}".format("S.No", "Car_Name", "Car_Quantity", "Price_in_Rupees_per_hour") + '\033[0m')
        for key, value in car_list.items():
            car_name, available_quantity, hour_price = value
            print("{:<5} {:<15} {:^15} {:^25}".format(key, car_name, available_quantity, hour_price))  
        print(f"\nTotal number of cars available currently:", sum)
    
    #It checks whether the car asked by customer is present in the car inventory.
    def check_car_in_inventory(self, car_name):
        car_found = 0 
        for car in car_list.values():
            if car[0] == car_name:
                car_found = 1
                return car_found
        if car_found == 0:
            return car_found
    
    #Function to find the price of any particular car.
    def find_price_of_car(self, car_name):
        for car in car_list.values():
            if car[0] == car_name:
                return car[2]
    
    #Function to update the car inventory after successful renting a car.
    def update_car_inventory(self, car_requested, car_quantity): 
        inventory_update = 0
        car_found_flag = 0
        for car in car_list.values():
            if car[0] == car_requested:
                car_found_flag = 1
                if car[1] >= car_quantity and car[1] != 0: 
                    car[1] = car[1] - car_quantity
                    inventory_update = 1
                else:
                    print("Sorry. Requested quantity cannot be rented")
        if car_found_flag == 0:
            print("Requested car not available")
        if inventory_update == 1:
            print("\nUpdated car inventory after booking: --") 
            obj_rented_cars = RentedCars() 
            obj_rented_cars.available_cars()
    
    #Function to generate the bill for a customer based on its registered name and phone number used during renting a car.
    def billing(self, customer_rental_list, name, phone_no):
        obj_rented_car = RentedCars()
        total_bill = 0
        customer_found = 0
        for item in customer_rental_list:
            if item[0] == name:
                if item[1] == phone_no:
                    customer_found = 1
                    price = obj_rented_car.find_price_of_car(item[2])
                    rented_duration = item[5] - item[4]
                    total_bill = int(item[3]) * item[5] * item[4] * price
                return total_bill, item[2], item[3], item[4], item[5] , price
        if customer_found != 1:
            return [0]*6 
    
    #Function to check whether requested car quantity by customer is less than or equal to available quantity in car inventory list.
    def check_car_count_availability(self, car_name, car_quantity):
        for car in car_list.values():
            if car[0] == car_name:
                if car[1] >= car_quantity:
                    return True
                else: 
                    return False
        
```

```python
class Customer(RentedCars): 
    '''
    Customer class for customer specific operation like requesting for car, 
    creating customer car rental record, booking list.
    '''
    
    obj_rented_cars = RentedCars()
    global customer_rental_list 
    customer_rental_list = []
    
    #Customer Constructor
    def __init__(self, name, phone_no):
        self.name = name
        self.phone_no = phone_no

    #Customer requesting for car to book 
    def car_request(self, car_requested, car_quantity):   
        obj_rented_cars.update_car_inventory(car_requested, car_quantity)

    #Function to create a customer car renting record for bill generation
    def customer_rental_record(self, car_requested, car_quantity, number_of_days, hours_per_day):
        customer_booking_list = [self.name, self.phone_no, car_requested, car_quantity, number_of_days, hours_per_day]
        customer_rental_list.append(customer_booking_list)
    
    #Function to fetch the list of customer record for billing
    def all_customer_booking_list(self):
        return customer_rental_list
```

```python
import sys
from datetime import datetime 

class CarRentingApp:
    '''
    CarRentingApp Class helps in renting car for the customer
    '''
    
    #Function to convert customer entered booking and return time to datetime format
    def time_conversion(self, booking_date, booking_time):
        book_date = booking_date.split("-")
        book_time = booking_time.split(":")
        booking_timestamp = datetime(int(book_date[2]), int(book_date[1]), int(book_date[0]), int(book_time[0]), int(book_time[1]), 0)
        return booking_timestamp
    
    #Function to extract days and hour difference between booking and return timestamp
    def days_and_hours_extract(self, booking_ending_timestamp, booking_starting_timestamp):
        time_delta = booking_ending_timestamp - booking_starting_timestamp
        days = time_delta.days
        hours =  math.ceil((time_delta.seconds)/3600)
        return (days+1), hours
    
    #This function will be called whenever program graceful termination is needed. 
    def exit_program(self):
        print("Exiting the program...")
        sys.exit(0)
    

#Main function
if __name__ == "__main__":
    
    #Car Renting App object creation
    obj_car_renting_app = CarRentingApp()
  
    #Rented Car object creation
    obj_rented_cars = RentedCars()
    
    #Display Menu 
    print("Menu:\n Enter '1' to Show Car Availability \n Enter '2' to Show All Rental modes \n Enter '3' to Book a Car \n Enter '4' to Generate Bill")
    choice = (int(input("\nEnter your Choice: ")))
    
    #Menu option 1: Display Available Cars
    if choice == 1: 
        print("\nHere is the list of all cars available with quantity and its pricing --\n")
        obj_rented_cars.available_cars()
        
    #Menu option 2: Display Rental Mode
    if choice == 2: 
        print("\nRental Mode Available:- \n Hourly \n Daily \n Weekly")
        
    #Menu option 3: Car Booking
    if choice == 3: 
        print("\n\n!!Welcome to car renting process!!\n")
        
        #Getting customer data for using app
        customer_name = input("Enter your name: ")
        customer_phone = input("Enter your phone number: ")
        obj_customer = Customer(customer_name, customer_phone)
        
        #Showing available car and quantity available for booking
        print("\nHere is the list of car available for booking -")
        obj_rented_cars.available_cars()
        
        #Taking car model from customer for booking 
        input_car = input("\nEnter the car for booking:")
        if obj_rented_cars.check_car_in_inventory(input_car) != 1:
            print(f"We currently do not have this {input_car} car.")
            try:
                print("Application terminating")
                obj_car_renting_app.exit_program()
            except Exception as e: 
                print("Exception occurred \n ", e)
                obj_car_renting_app.exit_program()
            finally:
                print("Thank you for using app.\n Rerun app to select car from available list.") 
        
        
        #Taking car quantity for the requested model from customer for booking
        input_car_quantity = int(input(f"Enter number of car for the selected brand {input_car}:"))
        #Checking if quantity of car is positive and less than available count
        if input_car_quantity > 0:
            if obj_rented_cars.check_car_count_availability(input_car, input_car_quantity) != 1:
                print(f"Requested quantity for selected {input_car} car is not avialable now. Kindly select different car from list.")
                try:
                    print("Application terminating")
                    obj_car_renting_app.exit_program()
                except Exception as e: 
                    print("Exception occurred ", e)
                    obj_car_renting_app.exit_program()
                finally:
                    print("Thank you for using app.\n Rerun app to select car from available list.")      
        else:
            print("Enter positive quantity")
        input_rental_mode = int(input(f"Select rental mode for car: 1) Hourly 2) Daily 3) Weekly"))
        
        #Different Rental Modes 
        #Hourly Mode
        if input_rental_mode == 1: 
            booking_start_date = input(f"You have selected car **  {input_car} **  with Quantity **  {input_car_quantity} **  \nPlease enter the booking start date in format dd-mm-yyyy (example: 03-02-2024)")
            booking_start_time = input("Please enter the starting time in format HH:MM (example: 15:00)")
            booking_end_time = input("Please enter the ending time in format HH:MM (example: 15:00)")
            booking_starting_timestamp = obj_car_renting_app.time_conversion(booking_start_date, booking_start_time)
            booking_ending_timestamp = obj_car_renting_app.time_conversion(booking_start_date, booking_end_time)
            days, hours_per_day = obj_car_renting_app.days_and_hours_extract(booking_ending_timestamp, booking_starting_timestamp)
            obj_customer.car_request(input_car, input_car_quantity)
            obj_customer.customer_rental_record(input_car, input_car_quantity, days, hours_per_day)
                    
        #Daily Mode
        elif input_rental_mode == 2: 
            booking_start_date = input(f"You have selected Car ** {input_car} ** with Quantity ** {input_car_quantity} ** \nPlease enter the booking start date in format dd-mm-yyyy (example: 03-02-2024)")
            booking_end_date = input("Please enter the booking end date in format dd-mm-yyyy (example: 04-02-2024)")
            booking_start_time = input("Please enter the starting time in format HH:MM (example: 15:00)")
            booking_end_time = input("Please enter the ending time in format HH:MM (example: 15:00)") 
            booking_starting_timestamp = obj_car_renting_app.time_conversion(booking_start_date, booking_start_time)
            booking_ending_timestamp = obj_car_renting_app.time_conversion(booking_end_date, booking_end_time)
            daily_days, hours_per_day = obj_car_renting_app.days_and_hours_extract(booking_ending_timestamp, booking_starting_timestamp)
            obj_customer.car_request(input_car, input_car_quantity)
            obj_customer.customer_rental_record(input_car, input_car_quantity, daily_days, hours_per_day)
            
        #Weekly Mode
        elif input_rental_mode == 3: 
            booking_start_date = input(f"You have selected car **  {input_car} **  with Quantity **  {input_car_quantity} ** \nPlease enter the booking start date in format dd-mm-yyyy (example: 03-02-2024)")
            booking_end_date = input("Please enter the booking end date in format dd-mm-yyyy (example: 04-02-2024)")
            booking_week_days = input("Enter day of the week. (Example, for Sunday and Tuesday enter '0, 2' comma seperated)")
            list_week_days = booking_week_days.split(",")
            number_of_days_in_week = len(list_week_days)
            booking_start_time = input("Please enter the starting time in format HH:MM (example: 15:00)")
            booking_end_time = input("Please enter the ending time in format HH:MM (example: 15:00)") 
            booking_starting_timestamp = obj_car_renting_app.time_conversion(booking_start_date, booking_start_time)
            booking_ending_timestamp = obj_car_renting_app.time_conversion(booking_end_date, booking_end_time)
            days, hours_per_day = obj_car_renting_app.days_and_hours_extract(booking_ending_timestamp, booking_starting_timestamp)
            weekly_mode_days = int(days / 7) * number_of_days_in_week
            obj_customer.car_request(input_car, input_car_quantity)
            obj_customer.customer_rental_record(input_car, input_car_quantity, weekly_mode_days, hours_per_day)
        else: 
            print("Wrong choice entered")
    
    #Generate bill for the customer
    if choice == 4: 
        print("\n\n!!Welcome to car rental bill generation!!")
        customer_name = input("Enter your name: ")
        customer_phone = input("Enter your phone number: ")
        obj_customer = Customer(customer_name, customer_phone)
        all_customer_booking_list = obj_customer.all_customer_booking_list()
        print("\nBased on the details enteried. Here is your bill -\n")
        total_bill, rented_car, rented_unit, days, hours_per_day, price_per_hour  = obj_customer.billing(all_customer_booking_list, customer_name, customer_phone)
        print(f"{customer_name} your total bill is - " +  '\033[1m' + "Rs" + str(total_bill) + '\033[0m')
        print("\n........Detailed Bill.........\n")
        print('\033[1m' + "{:<15} {:<15} {:<20} {:<20} {:<20} | {:<15}".format("Rented Car", "Quantity", "Rented_for_days", "Hours_per_day", "Price(per_hour)", "Total Price(in Rupee)") + '\033[0m')
        print("{:<15} {:<15} {:<20} {:<20} {:<20} | {:<15}".format(rented_car, rented_unit, days, hours_per_day,  price_per_hour, total_bill))
             
    
```
